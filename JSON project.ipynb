```python
import json
import pandas as pd
import numpy as np
from pandas.io.json import json_normalize
```

```python
df = pd.read_json('/Users/TCummings03/Desktop/data_wrangling_json/data/world_bank_projects.json')
```

```python
with open('/Users/TCummings03/Desktop/data_wrangling_json/data/world_bank_projects.json') as json_file:
    json_data = json.load(json_file)
```

```python
#1. Find the 10 countries with the most projects

Countries = df.countryshortname.value_counts()
print(Countries.nlargest(10))
```

```python
#2. Find the top 10 mjor project themes: Turn themes column into a dataframe and get top ten themes
datanormalized = json_normalize(json_data, 'mjtheme_namecode')

dfn = pd.DataFrame(datanormalized)

counts = dfn.code.value_counts()
```
